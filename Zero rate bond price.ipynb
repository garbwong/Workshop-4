```python
# Valuing a Zero - Coupon Bond with
# Constant Continuous Short Rate

#
from numpy import *
from matplotlib.pyplot import *

# parameters
r = [0.01 ,0.012 ,0.015 ,0.018 ,0.020]
T = 5.0
# valuation function
def BtT (r,t,T):
    return exp (-r*(T-t))*100
```

```python
# graphical analysis
t = linspace (0.0,T,10) 
for i in r:
    plot(t, BtT (i,t,T))
grid( True ); xlabel ('time '); ylabel ('bond price ')
```

```python
# #Valuation of (Zero - Coupon -) Bonds in the CIR Model

# Parameters
T=5.0;r0=0.01; kappa_r =0.140; theta_r =0.047; sigma_r =0.10

# ZCB Valuation Formula

def gamma ( kappa_r , sigma_r ):
    return sqrt ( kappa_r **2+2* sigma_r **2)

def b1( alpha ):
    r0 , kappa_r , theta_r , sigma_r ,t,T = alpha
    g = gamma ( kappa_r , sigma_r )
    return ((( 2*g* exp (( kappa_r +g )*(T-t)/2 ))/(2*g+( kappa_r +g )*( exp (g*(T-t))-1 )))**( 2* kappa_r * theta_r / sigma_r **2))

def b2( alpha ):
    r0 , kappa_r , theta_r , sigma_r ,t,T = alpha
    g = gamma ( kappa_r , sigma_r )
    return ((2*( exp (g*(T-t))-1 ))/(2*g+( kappa_r +g )*( exp (g*(T-t))-1 )))

def B( alpha ):
    b_1 = b1( alpha ); b_2 = b2( alpha )
    r0 , kappa_r , theta_r , sigma_r ,t,T = alpha
    Ert = theta_r + exp (- kappa_r *t )*( r0 - theta_r )
    return b_1 * exp (- b_2 * Ert )
```

```python
# Graphical Analysis
time = arange (0.0,5.5,0.5)
Bt = []
for t in time :
    alpha =[r0 , kappa_r , theta_r , sigma_r ,t,T]
    Bt. append (B( alpha ))
plot (time ,Bt ,'b')
plot (time ,Bt ,'ro ')
grid ( True )
xlabel ('years ')
ylabel ('CIR85 : B(t,T)')
```
