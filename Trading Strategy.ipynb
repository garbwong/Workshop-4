```python
#Groupby Cumsum Cumcount Reset Counter learned from Wallet T
```

```python
import pandas as pd
import numpy as np
```

```python
# read Excel file 'Trading Strategy.xlsx'
table = pd.read_excel('Trading Strategy.xlsx')
Direction = table['Direction'] # read the table
Direction
```

```python
condNoTrend = (Direction.shift(1).isnull()) | (abs(Direction + Direction.shift(1))<=1)# set the condition of no trend
table['Trend'] = np.where(condNoTrend,0,Direction) #setup a new column and check the condition. Tf true = 0, otherwise = Dir
Trend = table['Trend']

table #table inlcudes the columns of Direction and Trend now
```

```python
table['AccummDir'] = table.groupby((Trend==0).cumsum()).cumcount()
#cumsum will count if Trend =0, otherwise not count
#cumcount will count if cumsum do not count, otherwise not count

table
```
