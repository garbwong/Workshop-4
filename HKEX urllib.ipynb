```python
import bs4 as bs
import urllib.request
```

```python
sauce = urllib.request.urlopen('https://www.hkex.com.hk/eng/index.htm').read()
```

```python
soup = bs.BeautifulSoup(sauce, 'lxml')
```

```python
print(sauce)
```

```python
print(soup.title)
```

```python
print(soup.title.text)
```

```python
print(soup.p)
```

```python
print(soup.find_all('p'))
```

```python
for paragraph in soup.find_all('p'):
    print(paragraph.text) #string
```

```python
for url in soup.find_all('a'):
    print(url.text)
```

```python
for url in soup.find_all('a'):
    print(url.get('href'))
```
